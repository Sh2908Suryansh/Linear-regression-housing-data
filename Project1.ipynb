```python
df.dropna(subset=['zip_code'], inplace=True)
df['resarea'] = df['resarea'].fillna(df['resarea'].mean())
df['comarea'] = df['comarea'].fillna(df['comarea'].mean())
df['numfloors'] = df['numfloors'].fillna(df['numfloors'].mean())
df['latitude'] = df['latitude'].fillna(df['latitude'].mean())
df['longitude'] = df['longitude'].fillna(df['longitude'].mean())
df['landuse'] = df['landuse'].fillna(df['landuse'].mean())
```

# Regression

### As data is very big we will scale the data 

```python
features = ['block', 'borough_x', 'yearbuilt', 'longitude']
x = df[features].values
y = df['sale_price'].values.reshape(-1, 1)

xmean = x.mean(axis=0)                    #axis = 0 meANS COLUMN VISE
ymean = y.mean()
xstd = x.std(axis = 0)
ystd = y.std()

x_scaled = (x - xmean) /s xstd     # Z score 
y_scaled = (y - ymean) / ystd
x_scaled = np.c_[np.ones(x_scaled.shape[0]), x_scaled]

def linear_regression(X, y, step=0.01, tries=1000):
    m, n = X.shape
    weights = np.zeros((n, 1)) 
    cost_history = []
    for i in range(tries):
        predictions = np.dot(X, weights)
        error = predictions - y
        cost = np.mean(error ** 2)
        cost_history.append(cost)
        gradients = (2/m) * np.dot(X.T, error)
        weights = weights - (step * gradients)
    
    return weights, cost_history

final_weights, history = linear_regression(x_scaled, y_scaled, step=0.01, tries=1000)

plt.figure(figsize=(8, 5))
plt.plot(range(1000), history, color='red', linewidth=2)
plt.title('Cost History: Error Reduction Over Time', fontsize=14)
plt.xlabel('Epochs (Number of Iterations)', fontsize=12)
plt.ylabel('Mean Squared Error (Cost)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
